```python

import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings

warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df = pd.read_csv("./src/CarPrice_Assignment.csv")
df.head()
```

```python
df.columns
```

```python
selected_columns = [
    'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation',
    'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight',
    'enginetype', 'cylindernumber', 'enginesize', 'boreratio', 'stroke',
    'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg',
    'price'
]

df_train = df[selected_columns]
```

```python
cat_val = df_train.select_dtypes(include=['object']).columns
num_val = df_train.select_dtypes(include=['int', "float"]).columns

print(cat_val,'\n',num_val)
```

```python
df_train = df[selected_columns]

df_encoded = pd.get_dummies(df_train, drop_first=True)

target = 'price'
features = [col for col in df_encoded.columns if col != target]

X_train, X_test, y_train, y_test = train_test_split(
    df_encoded[features], df_encoded[target], test_size=0.2, random_state=42
)
```

```python
models = {
    'Support Vector Machine': SVR(kernel='rbf'),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'Bayesian Ridge': BayesianRidge(),
    'Linear Regression': LinearRegression()
}
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MAE': mae, 'R2': r2}

    # Plot results
    plt.figure(figsize=(12, 6))
    plt.plot(range(len(y_test)), y_test.values, label='Actual Price', color='blue')
    plt.plot(range(len(y_test)), y_pred, label='Predicted Price', color='red', linestyle='dashed')
    plt.title(f"Car Price Prediction using {name}")
    plt.xlabel("Sample Index")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

results
```
